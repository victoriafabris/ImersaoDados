# **Importando biblioteca e dados**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'} #'dicionario'
dados.rename(columns=mapa, inplace=True) #dados sendo modificados no local (em dados)
cod_compostos = dados['composto'].value_counts().index[0:5] #contagem dos 5 maiores valores em composto no dataset
```

# **Desafio 1** 
Neste desafio devemos ordenar o gráfico countplot, ou seja, ordenar as categorias do gráfico por contagem. 

Pode-se entender melhor sobre o gráfico countplot lendo a sua documentação no seguinte link: https://seaborn.pydata.org/generated/seaborn.countplot.html

O gráfico abaixo está fora de ordem:


```python
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 compostos')
plt.show()
```

```python
sns.set()
plt.figure(figsize=(8, 6))
dado_filtrado = dados.query('composto in @cod_compostos') #filtrando: pegando apenas os compostos presentes em cod_compostos
ax = sns.countplot(x = 'composto', data = dado_filtrado, order = dado_filtrado['composto'].value_counts().index, palette = ["#E0FFFF", "#F5F5DC", "#FDF5E6", "#FFE4E1", "#E6E6FA"])
ax.set_title('Top 5 compostos')
plt.show()
```

# **Desafio 2**
Para o desafio 2 devemos melhorar a visualização do gráfico acima alterando o tamanho da fonte. 

Este parâmetro é alterado pela biblioteca `seaborn` na função `seaborn.set()` ( `set_theme()` ), utilizando o parâmetro `font_scale`. Também podemos alterar o tamanho do título utilizando o parâmetro `fontsize` na função `set_title()` da bibliteca `matplotlib`.

```python
sns.set(font_scale = 1.25) 
plt.figure(figsize = (8, 6))
dado_filtrado = dados.query('composto in @cod_compostos') #filtrando: pegando apenas os compostos presentes em cod_compostos
ax = sns.countplot(x = 'composto', data = dado_filtrado, order = dado_filtrado['composto'].value_counts().index, palette = ["#E0FFFF", "#F5F5DC", "#FDF5E6", "#FFE4E1", "#E6E6FA"])
ax.set_title('Top 5 compostos', fontsize = 25)
plt.show()
```

# **Desafio 6**
 Fazer outras análises com o boxplot e até com o histograma, ambas funções da biblioteca `seaborn`.

 Primeiro, vamos criar a variável descricao contendo a descrição de todos os elementos das colunas g-0 até g-771 do dataset `dados`. Fazemos isso utilizando a função `loc[]` do Pandas.

```python
descricao = dados.loc[:,'g-0':'g-771'].describe()
descricao
```

Vamos plotar gráficos de algumas categorias descritivas desses valores. Para isso devemos **transpor** as colunas com as linhas, para podermos acessar as categorias.

```python
descricao = dados.loc[:,'g-0':'g-771'].describe().T
descricao
```

Fazemos agora a plotagem de um gráfico boxplot para a categoria `mean` (média) e também um histograma para a mesma categoria.

```python
fig, axes = plt.subplots(1, 2, figsize = (15,5))
sns.boxplot(x = 'mean' , data = descricao, ax = axes[0], color = "#D8BFD8")
sns.histplot(x = 'mean', data = descricao, bins = 10, ax = axes[1], color = "#D8BFD8")
plt.show()
```

O boxplot começa sempre no valor mínimo da base de dados e termina no valor máximo, assim como o histograma. 

No gráfico boxplot podemos observar que a o valor mínimo de mean se encontra em aproximadamente -0.5, podemos ver este valor exibido no histograma ao lado marcando este valor na sua primeira barra. O mesmo se repete para o valor máximo.

A caixa representa os valores centrais da base de dados, onde no histograma essa parte é representada pelas barras mais altas, ou seja, dados com maior frequência. 

No gráfico boxplot podemos observar que este valor se encontra em aproximadamente 0.0, exato valor onde a barra mais alta do histograma se encontra.

No boxplot, quando se tem um outlier, valores atípicos, facilmente conseguimos visualizá-los através do asterisco apresentado no boxplot. Isso pode ser observado no boxplot a direita na figura.

No boxplot podemos notar que há **outliers** (valores discrepantes nos dados), são exibidos por meio dos asteriscos. 

O histograma, quando há outliers, apresenta não mais uma curva normal centralizada, mas mais estendida para um lado do que para o outro, ou seja, uma **curva normal assimétrica**. Como observamos que há mais asteriscos no lado esquerdo do boxplot, o histograma fica mais estendido para a esquerda.
