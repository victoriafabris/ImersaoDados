# **Importando biblioteca e dados**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'} #'dicionario'
dados.rename(columns=mapa, inplace=True) #dados sendo modificados no local (em dados)
cod_compostos = dados['composto'].value_counts().index[0:5] #contagem dos 5 maiores valores em composto no dataset
```

# **Desafio 1** 
Neste desafio devemos ordenar o gr√°fico countplot, ou seja, ordenar as categorias do gr√°fico por contagem. 

Pode-se entender melhor sobre o gr√°fico countplot lendo a sua documenta√ß√£o no seguinte link: https://seaborn.pydata.org/generated/seaborn.countplot.html

O gr√°fico abaixo est√° fora de ordem:


```python
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 compostos')
plt.show()
```

```python
sns.set()
plt.figure(figsize=(8, 6))
dado_filtrado = dados.query('composto in @cod_compostos') #filtrando: pegando apenas os compostos presentes em cod_compostos
ax = sns.countplot(x = 'composto', data = dado_filtrado, order = dado_filtrado['composto'].value_counts().index, palette = ["#E0FFFF", "#F5F5DC", "#FDF5E6", "#FFE4E1", "#E6E6FA"])
ax.set_title('Top 5 compostos')
plt.show()
```

# **Desafio 2**
Para o desafio 2 devemos melhorar a visualiza√ß√£o do gr√°fico acima alterando o tamanho da fonte. 

Este par√¢metro √© alterado pela biblioteca `seaborn` na fun√ß√£o `seaborn.set()` ( `set_theme()` ), utilizando o par√¢metro `font_scale`. Tamb√©m podemos alterar o tamanho do t√≠tulo utilizando o par√¢metro `fontsize` na fun√ß√£o `set_title()` da bibliteca `matplotlib`.

```python
sns.set(font_scale = 1.25) 
plt.figure(figsize = (8, 6))
dado_filtrado = dados.query('composto in @cod_compostos') #filtrando: pegando apenas os compostos presentes em cod_compostos
ax = sns.countplot(x = 'composto', data = dado_filtrado, order = dado_filtrado['composto'].value_counts().index, palette = ["#E0FFFF", "#F5F5DC", "#FDF5E6", "#FFE4E1", "#E6E6FA"])
ax.set_title('Top 5 compostos', fontsize = 25)
plt.show()
```

# **Desafio 6**
 Fazer outras an√°lises com o boxplot e at√© com o histograma, ambas fun√ß√µes da biblioteca `seaborn`.

 Primeiro, vamos criar a vari√°vel descricao contendo a descri√ß√£o de todos os elementos das colunas g-0 at√© g-771 do dataset `dados`. Fazemos isso utilizando a fun√ß√£o `loc[]` do Pandas.

```python
descricao = dados.loc[:,'g-0':'g-771'].describe()
descricao
```

Vamos plotar gr√°ficos de algumas categorias descritivas desses valores. Para isso devemos **transpor** as colunas com as linhas, para podermos acessar as categorias.

```python
descricao = dados.loc[:,'g-0':'g-771'].describe().T
descricao
```

Fazemos agora a plotagem de um gr√°fico boxplot para a categoria `mean` (m√©dia) e tamb√©m um histograma para a mesma categoria.

```python
fig, axes = plt.subplots(1, 2, figsize = (15,5))
sns.boxplot(x = 'mean' , data = descricao, ax = axes[0], color = "#D8BFD8")
sns.histplot(x = 'mean', data = descricao, bins = 10, ax = axes[1], color = "#D8BFD8")
plt.show()
```

O boxplot come√ßa sempre no valor m√≠nimo da base de dados e termina no valor m√°ximo, assim como o histograma. 

No gr√°fico boxplot podemos observar que a o valor m√≠nimo de mean se encontra em aproximadamente -0.5, podemos ver este valor exibido no histograma ao lado marcando este valor na sua primeira barra. O mesmo se repete para o valor m√°ximo.

A caixa representa os valores centrais da base de dados, onde no histograma essa parte √© representada pelas barras mais altas, ou seja, dados com maior frequ√™ncia. 

No gr√°fico boxplot podemos observar que este valor se encontra em aproximadamente 0.0, exato valor onde a barra mais alta do histograma se encontra.

No boxplot, quando se tem um outlier, valores at√≠picos, facilmente conseguimos visualiz√°-los atrav√©s do asterisco apresentado no boxplot. Isso pode ser observado no boxplot a direita na figura.

No boxplot podemos notar que h√° **outliers** (valores discrepantes nos dados), s√£o exibidos por meio dos asteriscos. 

O histograma, quando h√° outliers, apresenta n√£o mais uma curva normal centralizada, mas mais estendida para um lado do que para o outro, ou seja, uma **curva normal assim√©trica**. Como observamos que h√° mais asteriscos no lado esquerdo do boxplot, o histograma fica mais estendido para a esquerda.
